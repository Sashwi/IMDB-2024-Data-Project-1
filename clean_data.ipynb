```python
import re

# Function to clean the 'Voting' column
def clean_voting(voting):
    try:
        voting = str(voting).strip("() ").upper().replace(",", "")
        if "K" in voting:
            return int(float(voting.replace("K", "")) * 1000)
        elif "M" in voting:
            return int(float(voting.replace("M", "")) * 1_000_000)
        elif voting.isdigit():
            return int(voting)
        else:
            return None
    except:
        return None

# Function to convert 'Duration' into minutes
def duration_to_minutes(duration):
    h = m = 0
    h_match = re.search(r"(\d+)h", duration)
    m_match = re.search(r"(\d+)m", duration)
    if h_match:
        h = int(h_match.group(1))
    if m_match:
        m = int(m_match.group(1))
    return h * 60 + m

# Clean the data
df['Voting'] = df['Voting'].apply(clean_voting)
df['Duration (min)'] = df['Duration'].apply(duration_to_minutes)
df['Title'] = df['Title'].apply(lambda x: re.sub(r"^\d+\.\s*", "", x).strip())
df['Genre'] = df['genre'].str.capitalize()

# Drop old genre column
df.drop(columns=['genre'], inplace=True)

# Drop duplicates
df.drop_duplicates(inplace=True)

# Drop rows with essential nulls
df.dropna(subset=['Title', 'Rating', 'Voting', 'Duration (min)', 'Genre'], inplace=True)

# Final data types check
df.dtypes
```
